```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("superstore_level2.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
df['Profit'] = df['Profit'].fillna(0)
```

```python
df = df.dropna()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Category'] = df['Category'].str.strip()
```

```python
df['Region'] = df['Region'].str.strip()
```

```python
top_states = df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
print(top_states)
```

```python
category_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
```

```python
print(category_profit)
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
print(region_sales)
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
```

```python
print(monthly_sales)
```

```python
sub_profit = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)
```

```python
print(sub_profit.head())
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Sales', data=df)
plt.title("Sales by Category")
plt.show()
```

```python
monthly_sales.plot(kind='line', figsize=(10,5))
plt.title("monthly sales trend")
plt.ylabel("sales")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Sales'], bins=30, kde=True)
plt.title("sales distribution")
plt.xlabel("sales")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title("profit vs sales")
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()

segment_sales.plot(kind='bar', figsize=(10,5))
plt.title("Sales by Customer Segment")
plt.ylabel("Sales")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['Sales','Profit','Discount','Quantity']].corr(), 
            annot=True, cmap='coolwarm')

plt.title("correlation heatmap")
plt.show()
```

```python
region_sales.plot(kind='bar', figsize=(10,5))
plt.title("region wise sales")
plt.show()
```

```python
top_sub = df.groupby('Sub-Category')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
top_sub.plot(kind='bar')
plt.title("profit by sub category")
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(10,10))
region_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title("region wise sales share")
plt.ylabel("")
plt.show()
```

## Insight Explanation

### 1. Technology Category Generates the Highest Revenue
From the category analysis, the **Technology category shows the highest total sales**, indicating strong demand for products such as phones, accessories, and machines. This suggests technology products contribute significantly to overall business revenue.

### 2. Certain States Dominate Sales Performance
The **top 5 states contribute a large portion of total sales**, showing that business performance is concentrated in a few geographic areas. These states represent strong markets where customer demand and purchasing activity are higher.

### 3. Western Region Shows the Highest Sales
The regional analysis indicates that the **West region generates the highest revenue compared to other regions**. This suggests stronger market penetration and higher customer purchasing power in this region.

### 4. Some Sub-Categories Generate Higher Profit
Profit analysis shows that **a few sub-categories generate significantly higher profits than others**, meaning these products have better margins and sales performance. Businesses should prioritize these profitable sub-categories to maximize revenue.

### 5. Sales Show a Clear Monthly Trend
The **monthly sales trend indicates fluctuations across different months**, showing that sales are not constant throughout the year. Certain months experience higher demand, which could be due to seasonal purchasing patterns or promotional periods.

### 6. Data Cleaning Improved Data Quality
Before analysis, the dataset contained **missing values and duplicate records**. Cleaning the data by removing duplicates and handling missing values improved the reliability of the analysis and ensured more accurate insights.
